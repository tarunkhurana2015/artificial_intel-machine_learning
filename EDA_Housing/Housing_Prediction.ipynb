```python
import warnings
warnings.filterwarnings('ignore')

#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("california_housing_data.csv")
df.head()
```

```python
df.info()
```

# the dataset contains 20,640 samples and 8 features;
# all features are numerical features encoded as floating number;
# there is no missing values.

```python
df.describe()
```

```python
avgRooms = df['total_rooms'] / df['households']
avgBedRooms = df['total_bedrooms'] / df['households']
avgHousehold = df['population'] / df['households']

df.insert(4, 'avgRooms', avgRooms)
df.insert(5, 'avgBedRooms', avgBedRooms)
df.insert(6, 'avgHousehold', avgHousehold)

df.drop(['total_rooms','total_bedrooms','households'], axis=1, inplace=True)

# in millions
df['median_house_value'] = df['median_house_value']/1000000.0

df.head(100)
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error


features = ['longitude', 'latitude', 'housing_median_age', 'avgRooms', 'avgBedRooms', 'avgHousehold','population','median_income']
X = df[features]
y = df['median_house_value']

y.head()
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Creating and fitting the linear regression model
lm = LinearRegression()
lm.fit(X_train, y_train)

# Making predictions on the test set
y_pred = lm.predict(X_test)

# Evaluating the model performance using mean squared error
mse1 = mean_squared_error(y_test, y_pred)
decimal_places = 7
mse2 = mse1 / (10 ** decimal_places)

# Round the result to the specified number of decimal places
mse = round(mse2-1, decimal_places)
print(f'Accuracy with Linear Regression: {mse}')
```
